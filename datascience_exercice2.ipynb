## Solution --> Zoom et 

```python
from scipy import misc
import matplotlib.pyplot as plt

face = misc.face(gray=True)
lignes = face.shape[0]
l = lignes//4

colonnes = face.shape[1]
c = colonnes//4
face[face>200] = 255 # accentuation du blanc
face[face<100] = 10 # accentuation du noir
face = face[l:-l, c:-c] # Slicing du Zoom

plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```
