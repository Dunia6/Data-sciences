# Exercice 2 ---> Zoom image

```python
from scipy import misc
import matplotlib.pyplot as plt

face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()

print(face)
print(face.shape)
```

## Solution

```python
from scipy import misc
import matplotlib.pyplot as plt

face = misc.face(gray=True)
lignes = face.shape[0]
l = lignes//4

colonnes = face.shape[1]
c = colonnes//4
face[face>190] = 255 # accentuation du blanc
face[face<100] = 50 # accentuation du noir
face = face[l:-l, c:-c] # Slicing du Zoom

plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

## Réduction de la qualité de l'image

```python
from scipy import misc
import matplotlib.pyplot as plt

face = misc.face(gray=True)
face = face[::15,::15] # slicing de la Dépréciation de la qualité de l'image

plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```
